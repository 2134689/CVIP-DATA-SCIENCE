# IMDB Movies Data Analysis

```python
#IMDB dataset link in kaggle : https://www,kaggle.com/datasets/PromptCloudHQ/imdb-data
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create the dataframe
df=pd.read_csv("IMDB-Movie-Data.csv")
df.head()
```

```python
#To get number of rows and columns 
df.shape
```

```python
df.dtypes
```

```python
#To get information about dataset
df.info()
```

```python
#Check missing values in the dataset
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
#To get missing value percentage
percentage_missing=df.isnull().sum()*100/len(df)
print(percentage_missing)
```

```python
#drop all missing values
df.dropna(axis=0)
```

```python
#Check for duplicate data
duplicate_data=df.duplicated().any()
print(duplicate_data)
```

```python
#drop to duplicate data
df=df.drop_duplicates() 
```

```python
#To get statistics about the dataframe
df.describe(include='all')
```

```python
#Genre analysis
df['Genre'].value_counts()
```

```python
seperate_genre='Action','Adventure','Sci-Fi','Drama','Comedy','Crime','Romance','Family','Fantasy','Mystery','Horror','Music','Musical','History','Biography'
for genre in seperate_genre:
    imdb = df['Genre'].str.contains(genre).fillna(False)
    print('The total no. of movies with ',genre,'=',len(df[imdb]))
    f, ax = plt.subplots(figsize=(10, 6))
    sns.countplot(x='Year', data=df[imdb], palette="Blues_d");
    plt.title(genre)
    compare_movies_rating = ['Runtime (Minutes)', 'Votes','Revenue (Millions)', 'Metascore']
```

```python
#Director analysis
df.columns
```

```python
#To get average rating for each director
df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
```

```python
df.Director.value_counts()[:15].plot.pie(autopct='%1.2f%%',figsize=(8,8))
plt.title('Top Ten Directors')
```

```python
#Actor analysis
df.Actors.value_counts()[:15].plot.pie(autopct='%1.2f%%',figsize=(8,8))
plt.title('Top Ten Actors')
```

```python
#Year analysis
df.columns
```

```python
#To get no. of movies per year
df["Year"].value_counts()
```

```python
sns.countplot(x='Year',data=df)
plt.show()
```

```python
#In which year was the highest average voting
df.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Votes',data=df)
plt.title("Average votes by year")
plt.show()
```

```python
#In which year there was the highest avergae revenue
df.columns
```

```python
df.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=df)
plt.title("Highest Average Revenue by year")
plt.show()
```

```python
#Rating analysis
df["Rating"].value_counts()
```

```python
#Rating Based on year
sns.stripplot(x="Year",y="Rating",data=df,color="Red")
```

```python
#Top 10 Movies from Rating analysis
top10=df.sort_values(['Rating'],ascending=False)
top10.head(10)
```

```python
#To display title of the movie having runtime >=180 minutes
df.columns
```

```python
df[df['Runtime (Minutes)']>=180]['Title']
```

```python
#To display top 10 lengthy movies Title and Runtime
top10=df.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']]\
.set_index('Title')
print(top10)
```

```python
sns.barplot(x='Runtime (Minutes)',y=top10.index,data=top10)
plt.show()
```

```python
#Finding most popular movie title(Highest Revenue)
df[df['Revenue (Millions)'].max()==df['Revenue (Millions)']]['Title']
```

```python
#Flop 3 movies with low rating
low10=df.query('(Rating>0) & (Rating <3.0)')
low10.head()
```
